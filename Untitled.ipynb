```python
# mon wrapper pour python 
api_key = 'N9C5WX01O3I1D7F5'

def get_day_adjusted_closing_price(
                                   function, 
                                   symbol, 
                                   outputsize, 
                                   datatype
                                  ): 
    key=api_key
    url='https://www.alphavantage.co/query?function={}&symbol={}&outputsize={}&datatype={}&apikey={}'.format(function, symbol, outputsize, datatype, key)
    response = requests.get(url)
    return response.json()


def get_quote_endpoint(symbol): 
    # hardcoded
    key=api_key
    function='GLOBAL_QUOTE'
    datatype='json' 
    # user_input
    symbol=symbol
    # get data 
    url='https://www.alphavantage.co/query?function={}&symbol={}&datatype={}&apikey={}'.format(function, symbol, datatype, key)
    response = requests.get(url)
    return response.json()
```
